# Exploratory Data Analysis - Terrorism

### Importing the Necessary Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Importing the Data

The data string is of latin1 encoding instead of UTF-8 encoding 

```python
df=pd.read_csv("globalterrorismdb_0718dist.csv", encoding='latin1', low_memory=False)
print("Data has been successfully imported.")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
for col in df.columns:
    print(col, end=", ")
```

## Cleaning and Filtering the Data

```python
df=df[["iyear","imonth","iday","country_txt","region_txt","provstate","city",
       "latitude","longitude","location","summary","attacktype1_txt","targtype1_txt",
       "gname","motive","weaptype1_txt","nkill","nwound","addnotes"]]
df.head()
```

```python
df = df.rename(columns={"iyear":"Year","imonth":"Month","iday":"Day","country_txt":"Country",
                   "region_txt":"Region","provstate":"Province/State","city":"City",
                   "latitude":"Latitude","longitude":"Longitude","location":"Location",
                   "summary":"Summary","attacktype1_txt":"Attack Type","targtype1_txt":"Target Type",
                   "gname":"Group Name","motive":"Motive","weaptype1_txt":"Weapon Type",
                   "nkill":"Killed","nwound":"Wounded","addnotes":"Add Notes"})
```

```python
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df["Killed"]=df["Killed"].fillna(0)
df["Wounded"]=df["Wounded"].fillna(0)
df["Casualty"]=df["Killed"]+df["Wounded"]
```

```python
df.describe()
```

### Observations:
1. The data consists of terrorist activities ranging from the year: 1970 to 2017
2. Maximum number of people killed in an event were: 1570
3. Maximum number of people wounded in an event were: 8191
4. Maximum number of total casualties in an event were: 9574

## Data Visualisation

### 1. Analysing attacks year-wise

1. Number of attacks in each year

```python
# get the counts of attacks year wise
attacks = df["Year"].value_counts(dropna=False)
# sort them on the basis of year
attacks = attacks.sort_index()
# convert to DataFrame and reset the index
attacks = attacks.to_frame().reset_index()
# rename the columns 
attacks = attacks.rename(columns={"index" : "Year", "Year" : "Attacks"})
# set the Year column as the index
attacks = attacks.set_index("Year")
```

```python
attacks.head()
```

Some conclusions: <br/>
Most number of attacks = 16903 in 2014 <br/>
Least number of attacks = 471 in 1971

```python
plt.figure(figsize=(15,6))
sns.barplot(x=attacks.index, y=attacks['Attacks'], color='blue', width=0.5, label='Attacks')
plt.xticks(rotation=90)
plt.legend()
plt.title("Timeline of Attacks")
plt.xlabel("Years")
plt.ylabel("Number of Attacks")
plt.show()
```

2. Total Casualties (Killed + Wounded) in each Year

```python
casualties=df[["Year","Casualty"]].groupby("Year").sum()
casualties.head()
```

```python
plt.figure(figsize=(15,6))
sns.barplot(x=casualties.index, y=casualties['Casualty'], color='red', width=0.5, label='Casualties')
plt.title("Year-wise Casualties")
plt.xlabel("Years")
plt.xticks(rotation=90)
plt.legend()
plt.ylabel("Number of Casualties")
plt.show()
```

3. Number of people killed each year

```python
killed = df[["Year","Killed"]].groupby("Year").sum()
killed.head()
```

4. Number of people wounded in each Region

```python
wounded=df[["Year","Wounded"]].groupby("Year").sum()
wounded.head()
```

```python
fig=plt.figure(figsize=(15,15))
ax0=fig.add_subplot(2,1,1)
ax1=fig.add_subplot(2,1,2)

# Killed plot
plt.figure(figsize=(15,15))
sns.barplot(x=killed.index, y=killed['Killed'], color='cornflowerblue', ax=ax0, label='Killed', width=0.5)
ax0.set_xticklabels(labels=killed.index, rotation=90)
ax0.legend()
ax0.set_title("People Killed in each Year")
ax0.set_xlabel("Years")
ax0.set_ylabel("Number of People Killed") 

# Wounded plot
plt.figure(figsize=(15,15))
sns.barplot(x=wounded.index, y=wounded['Wounded'], color='turquoise', ax=ax1, label='Wounded', width=0.5)
ax1.set_xticklabels(labels=wounded.index, rotation=90)
ax1.legend()
ax1.set_title("People Wounded in each Year")
ax1.set_xlabel("Years")
ax1.set_ylabel("Number of People Wounded")

plt.show()
```

### 2. Region wise Attacks

1. Distribution of Terrorist Attacks over Regions from 1970-2017

```python
# taking the cross product of the Region and Year attributes
reg=pd.crosstab(df.Year,df.Region)
reg.head()
```

```python
# using areaplot to find the region-wise number of attacks, each year
reg.plot(kind="area", stacked=False, alpha=0.5,figsize=(20,10))
plt.title("Region wise attacks",fontsize=20)
plt.xlabel("Years",fontsize=20)
plt.ylabel("Number of Attacks",fontsize=20)
plt.show()
```

2. Total Terrorist Attacks in each Region from 1970-2017 

```python
# taking transpose of the previous matrix
regt=reg.transpose()
# taking the total sum of all the attacks which is now sorted region-wise
regt["Total"]=regt.sum(axis=1)
# sort them in descending order
ra=regt["Total"].sort_values(ascending=False)
```

```python
plt.figure(figsize=(15,6))
sns.barplot(x=ra.index, y=ra.values, color='violet', label='Number of Attacks', width=0.75)
plt.legend()
plt.xticks(rotation=30)
plt.title("Total Number of Attacks in each region (1970-2017)")
plt.xlabel("Region")
plt.ylabel("Number of Attacks")
plt.show()
```

3. Total Casualties (Killed + Wounded) in each Region

```python
rc = df[["Region","Casualty"]].groupby("Region").sum()
rc = rc.sort_values(by='Casualty', ascending=False)
rc
```

```python
plt.figure(figsize=(15,6))
sns.barplot(x=rc.index, y=rc['Casualty'], color='orange', label='Casualty', width=0.75)
plt.legend()
plt.xticks(rotation=30)
plt.title("Region wise Casualties")
plt.xlabel("Regions")
plt.ylabel("Number of Casualties")
plt.show()
```

4. Killed in each Region

```python
rk=df[["Region","Killed"]].groupby("Region").sum()
rk = rk.sort_values(by="Killed",ascending=False)
rk
```

5. Wounded in each Region

```python
fig=plt.figure(figsize=(15,6))
ax0=fig.add_subplot(1,2,1)
ax1=fig.add_subplot(1,2,2)

# Killed
plt.figure(figsize=(15,6))
sns.barplot(x=rk.index, y=rk['Killed'], color='royalblue', ax=ax0, label='Killed', width=0.5)
ax0.set_xticklabels(labels=rk.index, rotation=90)
ax0.legend()
ax0.set_title("People Killed in each Region")
ax0.set_xlabel("Regions")
ax0.set_ylabel("Number of People Killed")

# Wounded
plt.figure(figsize=(15,6))
sns.barplot(x=rw.index, y=rw['Wounded'], color='lawngreen', ax=ax1, label='Wounded', width=0.5)
ax1.set_xticklabels(labels=rw.index, rotation=90)
ax1.legend()
ax1.set_title("People Wounded in each Region")
ax1.set_xlabel("Regions")
ax1.set_ylabel("Number of People Wounded")

plt.show()
```
